```python
def traverse(nd, kind="inorder"):
    items = []
    
    def inorder(nd):
        if nd is None: return
        inorder(nd.left)
        items.append(nd.val)
        inorder(nd.right)
    
    def preorder(nd):
        if nd is None: return
        items.append(nd.val)
        inorder(nd.left)
        inorder(nd.right)
        
    def postorder(nd):
        if nd is None: return
        inorder(nd.left)
        inorder(nd.right)
        items.append(nd.val)
            
    func = {
        "inorder": inorder,
        "preorder": preorder,
        "postorder": postorder
    }.get(kind)(nd)
    
    return items
```

```python
def invert(node):
    if node == None:
        return
    
    left = invert(node.left)
    right = invert(node.right)

    node.left = right
    node.right = left
    return node
```

```python
# level order
def level_order_traversal(node):
    if node is None:
        return
    
    items = []
    queue = []
    queue.append(node)
    
    while len(queue) > 0:
        currNode = queue[0]
        queue.remove(currNode)
        
        items.append(currNode.val)
        
        if currNode.left:
            queue.append(currNode.left)
        if currNode.right:
            queue.append(currNode.right)
    return items
```

```python
# max path sum
def maxPathSum(node):
    if node is None: return 0
    
    leftPath = maxPathSum(node.left)
    rightPath = maxPathSum(node.right)
    val = node.val + max(leftPath, rightPath)
    return val
```

```python
# min path sum
import math
def minPathSum(node):
    if node is None: return 0
    
    leftPath = minPathSum(node.left)
    rightPath = minPathSum(node.right)
    if leftPath == 0 and rightPath == 0:
        val = node.val
    elif leftPath == 0:
        val = node.val + rightPath
    elif rightPath == 0:
        val = node.val + leftPath
    else:
        val = node.val + min(leftPath, rightPath)
    return val
```

```python
def height(node):
    if node == None: return 0
    return 1 + max(height(node.left) , height(node.right))
```
